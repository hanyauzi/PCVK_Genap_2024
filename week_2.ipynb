```python
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np
img = cv.imread('/content/drive/MyDrive/pcvk/kitten.png')
plt.imshow(img)
```

img3 = cv.cvtColor(img, cv.COLOR_BGR2RGB): Ini mengonversi gambar dari format BGR (yang merupakan format default di OpenCV) menjadi format RGB (yang digunakan oleh matplotlib untuk menampilkan gambar). Ini penting karena jika gambar BGR langsung ditampilkan tanpa konversi, warna yang muncul akan salah (misalnya, biru akan terlihat merah).

```python
img2 = img
img3 = cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.imshow(img3)


```

```python
img_gray = cv.imread('/content/drive/MyDrive/pcvk/kitten.png', cv.IMREAD_GRAYSCALE)
plt.imshow(img_gray)
```

plt.imshow(img_gray): Menampilkan gambar img_gray. Secara default, plt.imshow()

cv.resize(..., (512, 1024)): Ini mengubah ukuran gambar hasil konversi ke resolusi 512 piksel lebar dan 1024 piksel tinggi. Fungsi ini meresize gambar dengan lebar dan tinggi yang baru

```python
img4 = cv.resize(cv.cvtColor (img,cv.COLOR_BGR2RGB), (512,1024))
plt. imshow(img4)
```

cv.flip(..., 0): Membalik gambar secara vertikal (flip atas-bawah). Argumen 0 menentukan arah flip vertikal, sedangkan 1 akan membalik secara horizontal.

```python
img5 = cv.flip(cv.cvtColor (img, cv.COLOR_BGR2RGB),0)
#simpan File image
#cv. imwrite('/content/drive/MyDrive/PCVK_MM/Images/mandrill2.tiff', img5)
#tampilkan plot dengan ukuran canvas yg Lebih besar
fig = plt.figure(figsize=(10,10))
#ubah-ubah ukuran (10,10) sesuai kebutuhan
ax = fig.add_subplot (111)
ax. imshow(img5)
```

black_img = np.zeros(shape=(512, 512, 3), dtype=np.int16):
* np.zeros(): Ini adalah fungsi dari library NumPy yang membuat sebuah array berisi nilai-nilai nol.
* shape=(512, 512, 3): Bentuk array yang dibuat adalah 512x512 piksel dengan 3 saluran (untuk RGB). Ini berarti gambar yang dibuat berukuran 512x512 dengan 3 channel warna, di mana semua pikselnya berwarna hitam (semua nilai saluran RGB = 0).
* dtype=np.int16: Tipe data setiap elemen dalam array adalah int16 (bilangan bulat 16-bit).
plt.imshow(black_img):
* Fungsi ini menampilkan array black_img sebagai gambar menggunakan matplotlib. Karena semua nilainya adalah nol, gambar yang ditampilkan akan berupa gambar hitam sepenuhnya.


```python
black_img = np.zeros(shape=(512,512,3),dtype=np. int16)
plt.imshow(black_img)
```

* cv.rectangle(): Fungsi OpenCV ini digunakan untuk menggambar persegi panjang pada gambar.
* black_img: Gambar hitam yang akan menjadi tempat persegi panjang.
* pt1=(384,0): Koordinat titik sudut kiri atas persegi panjang. Titik ini berada di x = 384, y = 0.
* pt2=(510,150): Koordinat titik sudut kanan bawah persegi panjang. Titik ini berada di x = 510, y = 150.
* color=(0,255,0): Warna garis persegi panjang dalam format RGB, dalam hal ini hijau (karena nilai hijau = 255).
* thickness=10: Ketebalan garis persegi panjang diatur menjadi 10 piksel.


```python
#perhatikan koordinat titik2 ptl dan pt2
cv.rectangle(black_img,pt1=(384,0),pt2=(510,150),color=(0,255,0),thickness=10)
plt. imshow(black_img)
```

* pt1=(200,200): Titik kiri atas dari persegi panjang, pada koordinat (200, 200).
* pt2=(300,300): Titik kanan bawah dari persegi panjang, pada koordinat (300, 300).
* color=(0,0,255): Warna garis persegi panjang dalam format RGB, di mana (0, 0, 255) adalah warna merah.
* thickness=15: Ketebalan garis persegi panjang diatur menjadi 15 piksel


```python
cv.rectangle(black_img,pt1=(200,200),pt2=(300,300), color=(0,0,255),thickness=15)
plt. imshow(black_img)
```

* cv.circle(): Fungsi ini digunakan untuk menggambar lingkaran pada gambar.
* black_img: Gambar hitam yang menjadi latar untuk lingkaran.
* center=(100,100): Koordinat titik pusat lingkaran, yang berada di (100, 100).
* radius=50: Jari-jari lingkaran adalah 50 piksel.
* color=(255,0,0): Warna lingkaran dalam format RGB, di mana (255, 0, 0) adalah warna biru (sesuai format BGR di OpenCV).
* thickness=8: Ketebalan garis lingkaran diatur menjadi 8 piksel. Jika Anda ingin menggambar lingkaran yang terisi penuh, Anda bisa menggunakan nilai -1 untuk parameter ini.


```python
cv.circle(black_img, center=(100,100), radius=50, color=(255,0,0), thickness=8)
plt. imshow(black_img)
```

* cv.line(): Fungsi ini digunakan untuk menggambar garis pada gambar.
* black_img: Gambar hitam yang menjadi latar tempat garis akan digambar.
* pt1=(0,0): Koordinat titik awal garis, yaitu di sudut kiri atas gambar.
* pt2=(512,512): Koordinat titik akhir garis, yaitu di sudut kanan bawah gambar (asumsi ukuran gambar adalah 512x512 piksel).
* color=(255,255,255): Warna garis dalam format RGB, di mana (255, 255, 255) adalah warna putih.
* thickness=5: Ketebalan garis diatur menjadi 5 piksel.


```python
cv. line (black_img,pt1=(0,0),pt2=(512,512), color=(255,255,255), thickness=5)
plt. imshow(black_img)
```

* cv.putText(): Fungsi ini digunakan untuk menambahkan teks pada gambar.
* black_img: Gambar hitam yang akan digunakan untuk menampilkan teks.
* text='Hello': Teks yang akan ditambahkan ke gambar.
* org=(10,500): Koordinat posisi titik dasar teks pada gambar, yaitu (10, 500). Titik ini adalah posisi di mana teks dimulai.
* fontFace=font: Jenis font yang digunakan untuk teks, yaitu FONT_HERSHEY_SIMPLEX.
* fontScale=4: Skala font yang menentukan ukuran teks. Nilai ini memperbesar ukuran font.
* color=(255,255,0): Warna teks dalam format RGB, di mana (255, 255, 0) adalah warna kuning.
* thickness=2: Ketebalan garis teks.
* lineType=cv.LINE_AA: Jenis garis anti-aliasing untuk membuat teks lebih halus

```python
font = cv. FONT_HERSHEY_SIMPLEX
cv.putText(black_img,text='Hello',org=(10,500),fontFace=font,fontScale=4,color=(255,255,0), thickness=2,lineType=cv.LINE_AA)
plt.imshow(black_img)
```

* np.zeros(): Fungsi ini membuat sebuah array dengan semua elemen diinisialisasi ke nol.
* shape=(512, 512, 3): Bentuk array adalah 512x512 piksel dengan 3 saluran (untuk RGB). Ini berarti gambar yang dibuat berukuran 512x512 dengan 3 saluran warna.
* dtype=np.int32: Tipe data setiap elemen dalam array adalah int32 (bilangan bulat 32-bit). Ini memungkinkan nilai warna dari setiap piksel untuk berada dalam rentang lebih besar dibandingkan int16 atau uint8, tetapi dalam konteks gambar, biasanya nilai warna berada dalam rentang 0 hingga 255.


```python
black_img2=np.zeros(shape=(512,512,3),dtype=np. int32)
plt. imshow(black_img2)
```

* np.array(): Fungsi ini digunakan untuk membuat array NumPy.
* [[100,300], [200,200], [400,300], [200,400]]: List dua dimensi yang berisi koordinat titik-titik vertikal. Setiap sub-list berisi dua angka yang mewakili koordinat x dan y dari satu titik.
* dtype=np.int32: Menetapkan tipe data elemen dalam array sebagai int32 (bilangan bulat 32-bit).

```python
vertices = np.array ([[100,300], [200,200], [400,300], [200,400]],dtype=np.int32)
vertices
```

* black_img2: Gambar hitam yang menjadi latar untuk poligon.
* [vertices]: List yang berisi array koordinat titik-titik poligon. Pastikan vertices didefinisikan dengan benar.
* isClosed=True: Menutup poligon dengan menghubungkan titik akhir ke titik awal.
* color=(255,0,0): Warna garis poligon dalam format RGB, di mana (255, 0, 0) adalah merah.
* thickness=5: Ketebalan garis poligon adalah 5 piksel.


```python
cv.polylines(black_img2, [vertices], isClosed=True, color=(255,0,0), thickness=5)
plt.imshow(black_img2)
```
